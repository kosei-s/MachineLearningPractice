```
import numpy as np

x = np.array([2, 3, 1])
t = np.array([20])
```

```
w1 = np.array([[3, 1, 2], [-2, -3, -1]])
b1 = np.array([0, 0])

w2= np.array([3, 2])
b2 = np.array([0])
```

```
u1 = w1.dot(x) + b1
h1 = 1. / (1 + np.exp(-u1))

y = w2.dot(h1) + b2

print(y)
```

```
dLdy = -2 * (t - y)

dydw2 = h1
```

```
dLdw2 = dLdy * dydw2

print(dLdw2)
```

```
dydh1 = w2
dh1du1 = h1 * (1 - h1)
du1dw1 = x

dLdu1 = dLdy * dydh1 * dh1du1
dLdu1 = dLdu1[None]
du1dw1 = du1dw1[None]

print(dLdu1.shape)
print(du1dw1.shape)

print(dLdu1.T.shape)

dLdw1 = (dLdu1.T).dot(du1dw1)

print(dLdw1)
```
